```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_parabola():
    # Define parameter 'a' (distance from vertex to focus)
    a = 1
    
    # Generate x values for the parabola x^2 = 4ay -> y = x^2 / 4a
    x = np.linspace(-5, 5, 400)
    y = x**2 / (4 * a)
    
    # Create the plot
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # 1. Plot the Parabola curve
    ax.plot(x, y, label=f'Parabola ($x^2 = 4ay$)', color='blue', linewidth=2)
    
    # 2. Plot the Focus (0, a)
    ax.plot(0, a, 'ro', markersize=8, label='Focus (0, a)')
    ax.text(0.2, a, f'Focus (0, {a})', fontsize=12, verticalalignment='bottom')
    
    # 3. Plot the Vertex (0, 0)
    ax.plot(0, 0, 'ko', markersize=8, label='Vertex (0, 0)')
    ax.text(0.2, 0, 'Vertex (0, 0)', fontsize=12, verticalalignment='top')
    
    # 4. Plot the Directrix (y = -a)
    ax.axhline(y=-a, color='green', linestyle='--', linewidth=2, label='Directrix ($y = -a$)')
    ax.text(2, -a - 0.3, f'Directrix y = -{a}', fontsize=12, color='green')
    
    # 5. Visual Proof of Definition (Distance to Focus = Distance to Directrix)
    # Let's pick a point P on the parabola, e.g., at x = 4
    px = 4
    py = px**2 / (4 * a) # y = 4
    
    # Plot Point P
    ax.plot(px, py, 'mo', markersize=8, label='Point P(x, y)')
    ax.text(px + 0.2, py, f'P({px}, {py})', fontsize=12)
    
    # Draw line from P to Focus
    ax.plot([px, 0], [py, a], 'r:', linewidth=1.5)
    ax.text(px/2, (py+a)/2 + 0.2, '$d_1$', fontsize=12, color='red')
    
    # Draw line from P to Directrix (perpendicular)
    ax.plot([px, px], [py, -a], 'g:', linewidth=1.5)
    ax.text(px + 0.1, (py-a)/2, '$d_2$', fontsize=12, color='green')

    # Formatting the graph
    ax.axhline(0, color='black', linewidth=1) # x-axis
    ax.axvline(0, color='black', linewidth=1) # y-axis
    ax.grid(True, linestyle=':', alpha=0.6)
    ax.set_aspect('equal')
    ax.set_xlim(-6, 6)
    ax.set_ylim(-3, 6)
    ax.set_title(f'Parabola Anatomy: Focus vs Directrix (a={a})', fontsize=14)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend(loc='upper left')

    plt.show()

if __name__ == "__main__":
    visualize_parabola()
```
